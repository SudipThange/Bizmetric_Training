> QUESTION 11:

>Ask user to enter the marks and define if the user got distinction, first class, second class, pass, Fails

>If marks are more than 80 output must be “You got distinction”

>For marks more than 60 output must be “You got First class”

>Marks more than 40 will display “You got second class”

>Marks more than or equal to 35 “PASS” Otherwise Fail

```python
def print_grades(marks):
    if marks >= 80:
        print("You got distinction!!")
    elif marks >= 60 and marks < 80:
        print("You got First Class!!")
    elif marks >= 40 and marks < 60:
        print("You got Second Class!!")
    elif marks >= 35 and marks < 40:
        print("Pass!!")
    else:
        print("Fail!!")

marks = float(input("ENTER THE MARKS?: "))
print_grades(marks)
```

```python
def display(old_sal, increment):
    print("-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-")
    print("OLD SALARY IS: ", old_sal)
    print("INCREMENT AMOUNT FOR {per}% IS: {val}".format(
        per=increment,
        val=old_sal / 100 * increment
    ))
    print("FINAL AMOUNT: ", old_sal + (old_sal / 100 * increment))
    print("-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-")

def filter(sal, rating): 
    if sal > 0 and rating in ('A', 'B', 'C', 'D'):

        if sal >= 2300000:
            if rating == 'A':
                display(sal, 7)
            elif rating == 'B':
                display(sal, 5)
            elif rating == 'C':
                display(sal, 4)
            else:
                display(sal, 0)

        elif sal >= 1500000 and sal < 2300000:
            if rating == 'A':
                display(sal, 8)
            elif rating == 'B':
                display(sal, 6)
            elif rating == 'C':
                display(sal, 4)
            else:
                display(sal, 0)

        elif sal >= 1000000 and sal < 1500000:
            if rating == 'A':
                display(sal, 14)
            elif rating == 'B':
                display(sal, 10)
            elif rating == 'C':
                display(sal, 8)
            else:
                display(sal, 6)

        elif sal >= 500000 and sal < 1000000:
            if rating == 'A':
                display(sal, 16)
            elif rating == 'B':
                display(sal, 12)
            elif rating == 'C':
                display(sal, 10)
            else:
                display(sal, 6)

    else:
        print("WRONG INPUT!!!")

    
def get_data():
    sal = float(input("ENTER THE SALARY?: "))
    rating = input("ENTER THE RATING?: ")
    return (sal, rating)

# main code #
sal, rating = get_data()
filter(sal, rating)
```

```python
from datetime import datetime

# ---------------- Utility Functions ----------------

def get_standard(standard_input):
    if 1 <= standard_input <= 4:
        return "1-4"
    elif 5 <= standard_input <= 8:
        return "5-8"
    elif 9 <= standard_input <= 10:
        return "9-10"
    else:
        return None


def convert(nb_input):
    if nb_input == 1:
        return "100 pages"
    elif nb_input == 2:
        return "200 pages"
    else:
        return nb_input


def get_quantity():
    return int(input("\nEnter the Quantity?: "))


def add_to_cart(cart, product, price, quantity):
    cart["Products"].append(product)
    cart["Price"].append(price)
    cart["Quantity"].append(quantity)


def print_header():
    print("=" * 100)
    print("{:^100}".format("BILL RECEIPT"))
    print("=" * 100)


def print_footer():
    print("=" * 100)


def print_items(title, cart):
    print("-" * 100)
    print(f"{title}")
    print("-" * 100)
    print("{:50} {:10} {:10}".format("Products", "Price", "Quantity"))

    for i in range(len(cart["Products"])):
        print("{:45} {:10} {:10}".format(
            cart["Products"][i],
            cart["Price"][i],
            cart["Quantity"][i]
        ))


def print_bill(total_amount, books_cart, notebook_cart):
    print("\n")
    print_header()

    if len(books_cart["Products"]) > 0:
        print_items("BOOKS", books_cart)

    if len(notebook_cart["Products"]) > 0:
        print_items("NOTEBOOKS", notebook_cart)

    print("-" * 100)
    print(f"Total Amount ===>> {total_amount}")
    print_footer()


def download_bill(bookscart, notebookscart, total_amount):
    try:
        file_path = (
            r"C:\Users\Sudip\Downloads\Bizmetric Training\Python\Bills\\"
            + datetime.now().strftime("%H_%M_%S") + ".txt"
        )

        with open(file_path, 'a') as file:
            file.write("=" * 100 + "\n")
            title = "BILL RECEIPT " + '(' +str(datetime.now()) + ')'
            file.write("{:^100}\n".format(title))
            file.write("=" * 100 + "\n")

            if bookscart["Products"]:
                file.write("BOOKS\n")
                file.write("-" * 100 + "\n")
                for i in range(len(bookscart["Products"])):
                    file.write(
                        f"{bookscart['Products'][i]}  "
                        f"{bookscart['Price'][i]}  "
                        f"{bookscart['Quantity'][i]}\n"
                    )

            if notebookscart["Products"]:
                file.write("\nNOTEBOOKS\n")
                file.write("-" * 100 + "\n")
                for i in range(len(notebookscart["Products"])):
                    file.write(
                        f"{notebookscart['Products'][i]}  "
                        f"{notebookscart['Price'][i]}  "
                        f"{notebookscart['Quantity'][i]}\n"
                    )

            file.write("\n" + "-" * 100 + "\n")
            file.write(f"TOTAL AMOUNT ===>> {total_amount}\n")

        print("BILL DOWNLOADED SUCCESSFULLY!")
        print("LOCATION:", file_path)

    except Exception as e:
        print("ERROR:", e)


# ---------------- Data ----------------

books_data = {
    '1-4': {'hindi': 60, 'marathi': 60, 'english': 80, 'science': 90, 'maths': 100},
    '5-8': {'hindi': 100, 'marathi': 100, 'english': 100, 'science': 120, 'maths': 140},
    '9-10': {'hindi': 150, 'marathi': 150, 'english': 150, 'science': 200, 'maths': 250}
}

notebooks_data = {
    '100 pages': {'square': 40, 'four_lines': 30, 'two_lines': 30, 'single_lines': 60, 'a4': 100},
    '200 pages': {'square': 70, 'four_lines': 50, 'two_lines': 50, 'single_lines': 100, 'a4': 180}
}

total_amt = 0
books_cart = {"Products": [], "Price": [], "Quantity": []}
notebook_cart = {"Products": [], "Price": [], "Quantity": []}


# ---------------- Menu ----------------

def menu():
    total_amt = 0

    while True:
        print("\n" + "*" * 50)
        print("{:^50}".format("STATIONERY MENU"))
        print("*" * 50)
        print("Enter 1: Buy Books")
        print("Enter 2: Buy Notebook")
        print("Enter 3: Exit")

        choice = int(input("ENTER YOUR CHOICE?: "))

        # -------- Books --------
        if choice == 1:
            standard_input = int(input("\nENTER THE STANDARD?: "))
            standard = get_standard(standard_input)

            if not standard:
                print("INVALID STANDARD!")
                continue

            while True:
                print("\nHindi / Marathi / English / Science / Maths")
                print("Exit")
                your_choice = input("ENTER YOUR CHOICE?: ").lower()

                if your_choice == "exit" or your_choice not in books_data[standard]:
                    break

                quantity = get_quantity()
                total_amt += books_data[standard][your_choice] * quantity

                add_to_cart(
                    books_cart,
                    your_choice.capitalize() + " Book",
                    books_data[standard][your_choice],
                    quantity
                )
                print("BOOK ADDED!")

        # -------- Notebooks --------
        elif choice == 2:
            while True:
                print("\n1: 100 Pages")
                print("2: 200 Pages")
                print("3: Exit")

                nb_choice = int(input("ENTER YOUR CHOICE?: "))

                if nb_choice == 3:
                    break

                nb_type = convert(nb_choice)

                while True:
                    print("\nSquare / Four_lines / Two_lines / Single_lines / A4")
                    print("Exit")
                    your_input = input("ENTER YOUR CHOICE?: ").lower()

                    if your_input == "exit" or your_input not in notebooks_data[nb_type]:
                        break

                    quantity = get_quantity()
                    total_amt += notebooks_data[nb_type][your_input] * quantity

                    add_to_cart(
                        notebook_cart,
                        your_input.capitalize().replace('_', ' ') + " Notebook",
                        notebooks_data[nb_type][your_input],
                        quantity
                    )
                    print("NOTEBOOK ADDED!")

        elif choice == 3:
            print("EXITING...")
            break


# ---------------- Run Program ----------------

menu()
print_bill(total_amt, books_cart, notebook_cart)
download_bill(books_cart, notebook_cart, total_amt)
```

```python
def create_using_loop(start, end):
    li = []
    
    for i in range(start, end + 1):
        if i % 2 != 0:
            li.append(i)
            
    return li

li1 = [i for i in range(1, 11) if i % 2 != 0]
print("ODD NUMBERS: ", li1)

li2 = create_using_loop(1, 10)
print("ODD NUMBERS: ", li2)
```

>38. Create a function to validate user first name/last name. User first name/last name should contain only characters. No special characters, numbers, space in name
>***

```python
def print_data(first_name, last_name):
    if first_name != "" and last_name != "":
        print(f"USER NAME IS: {first_name + ' ' + last_name} !!")
    else:
        print("WRONG INPUT!!")


def validate(str):
    try:
        if not str.isalpha() or str == "":
            raise Exception("WRONG INPUT!!")
        else:
            return True
    except Exception as e:
        print(e)
        return False


def get_data():
    first_name = input("ENTER THE NAME?: ")
    last_name = input("ENTRE THE LAST NAME?: ")
    
    if validate(first_name):
        first_name = first_name
        
        if validate(last_name):
            last_name = last_name
        else:
            return None
    else:
        return None
        
        
    return first_name, last_name


data = get_data()
if data is not None:
    first_name, last_name = data
    print_data(first_name, last_name)
```

>41. Create a empty dictionary and ask user to enter values as name, DOB, mobile number add all the details in dictionary with customer number as 1 for first time. If user try to enter another value, then number should increase as 2 with new details and previous values should not change.
>
>For example:
>
>{}
>
>{1:{name : "Sachin", "DOB": "21-06-1965" , "mobile": "1234123423"}}
>
>
>{1:{name : "Sachine", "DOB": "21-06-1965" , "mobile": "1234123423"},
>
>2: {name : "Sumedh", "DOB": "02-02-2002" , "mobile": "1234123433"}}
>***

```python
def validate_name(name):
    if name.isalpha():
        return True
    else:
        return False
        
def validate_dob(dob):
    if len(dob) != 10 or dob[2] != '-' or dob[5] != '-':
        return False

    day = int(dob[:2])
    month = int(dob[3:5])

    if (day >= 1 and day <= 31) and (month >= 1 and month <= 12):
        return True
    
    return False
 
        
def validate_mobile_number(mob_no):
    if mob_no.isdigit() and len(mob_no) == 10 and mob_no[0] != '0':
        return True
    return False

    
def get_data():
    name = int(input("Enter the name?: "))
    if not validate_name(name):
        print("WRONG INPUT FOR THE `Name`")
        return None
    
    dob = input("Enter the `Date of Birth` (DD-MM-YYYY)?: ")
    if not validate_dob(dob):
        print("WRONG INPUT FOR THE `Date of Birth`")
        return None
    
    mobile_number = input("ENTER THE `Mobile Number`?: ")
    if not validate_mobile_number(mobile_number):
        print("WRONG INPUT FOR THE `Mobile Number`")
        return None
    
    return name, dob, mobile_number
    
def add_user(users, name, dob, mobile_number):
    sr_no = len(users)
    
    sr_no += 1
    temp = {"name": name, "date of birth": dob, "mobile number": mobile_number}
    users.update({sr_no: temp})
    
    print("USER ADDED!!")
    
def print_details(users):
    for key in users:
        print('|| Sr.No:', key, end = " ")
        for k in users[key]:
            print(f"{k.capitalize()}: ", users[key][k], end = " || ")
        print()
```

```python
users = {}
```

```python
data = get_data()

if data is not None:
    name, dob, mobile_number = data
    add_user(users, name, dob, mobile_number)
    print_details(users)
```

```python
class AgeException(Exception):
    def __init__(self, msg):
        self.msg = msg
        
    def __str__(self):
        return f"{self.msg}"

def get_age():
    while True:
        try:
            age = input("ENTER THE AGE?: ")
            age = int(age)
            
            if age > 0 and age < 130:
                return f"AGE IS: {age}"
            else:
                raise AgeException("You entered the wrong age, Plese Enter Again....")
        except ValueError:
            print("Alphabets are not allowed....")
        except Exception as e:
            print(e)
        
print(get_age())
```
