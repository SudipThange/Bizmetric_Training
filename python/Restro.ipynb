```python
from datetime import datetime
import os

class Menu:
    starters = {
        1: {"code": "s01", "name": "Idli Dosa", "price": 200},
        2: {"code": "s02", "name": "Pav Bhaji", "price": 180},
        3: {"code": "s03", "name": "Paneer Tikka", "price": 250},
        4: {"code": "s04", "name": "Paneer 65", "price": 230},
        5: {"code": "s05", "name": "Kimchi", "price": 220},
        6: {"code": "s06", "name": "Korean Fried Chicken", "price": 350},
        7: {"code": "s07", "name": "Tteokbokki", "price": 300},
        8: {"code": "s08", "name": "Mandu", "price": 280},
    }

    dinner = {
        1: {"code": "d01", "name": "Butter Chicken", "price": 420},
        2: {"code": "d02", "name": "Paneer Butter Masala", "price": 380},
        3: {"code": "d03", "name": "Dal Tadka", "price": 240},
        4: {"code": "d04", "name": "Veg Biryani", "price": 300},
        5: {"code": "d05", "name": "Chicken Biryani", "price": 420},
        6: {"code": "d06", "name": "Garlic Naan", "price": 70},
        7: {"code": "d07", "name": "Jeera Rice", "price": 180},
        8: {"code": "d08", "name": "Korean Bibimbap", "price": 450},
        9: {"code": "d09", "name": "Korean BBQ Chicken", "price": 520},
        10: {"code": "d10", "name": "Kimchi Fried Rice", "price": 350},
    }

    desserts = {
        1: {"code": "ds01", "name": "Gulab Jamun", "price": 120},
        2: {"code": "ds02", "name": "Rasgulla", "price": 110},
        3: {"code": "ds03", "name": "Chocolate Brownie", "price": 180},
        4: {"code": "ds04", "name": "Ice Cream Sundae", "price": 160},
        5: {"code": "ds05", "name": "Mochi Ice Cream", "price": 220},
        6: {"code": "ds06", "name": "Honey Pancakes", "price": 250},
    }

    def get_quantity(self):
        while True:
            try:
                quantity = int(input("ENTER THE QUANTITY?: "))
                if quantity <= 0:
                    raise Exception("ENTERED WRONG VALUE TRY AGAIN...")
                return quantity
            except Exception as e:
                print(e)

    def convert_code(self, code, menu_di):
        return menu_di[code]

    def order(self, menu):
        res = []
        while True:
            try:
                code = int(input("Enter the choice?: "))

                if code == len(menu) + 1:
                    break

                if code < 1 or code > len(menu):
                    raise Exception("PLEASE ENTER THE CHOICE MENTIONED ABOVE...")

                quantity = self.get_quantity()

                # FIXED: prevent dictionary mutation
                item = self.convert_code(code, menu).copy()
                item["quantity"] = quantity

                res.append(item)
                print("ORDER ADDED!!")

            except ValueError:
                print("PLEASE ENTER A VALID CHOICE...")
            except Exception as e:
                print(e)

        if res:
            return res

    def show_menu(self, title, menu):
        print("*" * 100)
        print("|{:^98}|".format(title.upper()))
        print("*" * 100)

        for i in range(len(menu)):
            text = "> " + str(i + 1) + " " + menu[i + 1]["name"].capitalize()
            print("\t{}".format(text))
        print("\t{}".format("> " + str(len(menu) + 1) + " EXIT"))

        print("- " * 50)
        
class Orders:
    tables = [i for i in range(1, 16)]
    tables_reserved = []

    def validate_table_number(self):
        while True:
            try:
                table_number = int(input("ENTER TABLE NUMBER?: "))

                if table_number not in Orders.tables:
                    raise Exception("PLEASE ENTER BETWEEN 1 TO 15...")

                if table_number in Orders.tables_reserved:
                    raise Exception("TABLE IS ALREADY BOOKED, PLEASE WAIT...")

                return table_number

            except ValueError:
                print("PLEASE ENTER NUMBER ONLY...")
            except Exception as e:
                print(e)

    def take_order(self):
        try:
            table_number = self.validate_table_number()

            if table_number:
                Orders.tables_reserved.append(table_number)
                menu = Menu()

                your_orders = {
                    'order id': "ORD0" + str(table_number) + "_" +
                                str(datetime.now().strftime("%I-%M")),
                    'table number': table_number,
                    'date': str(datetime.now().strftime("%d-%b-%Y")),
                }

                while True:
                    print("-" * 100)
                    print("------> MENU <--------")
                    print("ENTER 1: FOR STARTERS")
                    print("ENTER 2: FOR DINNER")
                    print("ENTER 3: FOR DESSERTS")
                    print("ENTER 4: EXIT")

                    choice = int(input("ENTER YOUR CHOICE?: "))

                    if choice == 1:
                        menu.show_menu("starters", Menu.starters)
                        item = menu.order(Menu.starters)
                        if item:
                            your_orders["starters"] = item

                    elif choice == 2:
                        menu.show_menu("dinner", Menu.dinner)
                        item = menu.order(Menu.dinner)
                        if item:
                            your_orders["dinner"] = item

                    elif choice == 3:
                        menu.show_menu("desserts", Menu.desserts)
                        item = menu.order(Menu.desserts)
                        if item:
                            your_orders["desserts"] = item

                    elif choice == 4:
                        break

                    else:
                        raise Exception("PLEASE ENTER A VALID CHOICE...")

                if (your_orders.get("starters") or
                        your_orders.get("dinner") or
                        your_orders.get("desserts")):

                    response = input("PLEASE CONFIRM YOUR ORDERS [Y/N]?: ").lower()

                    if response not in ("y", "n"):
                        raise Exception("PLEASE ONLY ENTER YES OR NO...")

                    if response == "y":
                        return your_orders
                    else:
                        Orders.tables_reserved.remove(table_number)
                        raise Exception("ORDER CANCELLED...")

                else:
                    Orders.tables_reserved.remove(table_number)
                    raise Exception("YOU HAVEN'T ORDERED ANYTHING...")

        except ValueError:
            print("PLEASE ENTER NUMBER ONLY...")
        except Exception as e:
            print(e)

    def deliver_order(self):
        pass

    def print_bill(self, orders):
        try:
            print("=" * 100,)
            print("|{:^98}|".format("BILL"))
            print("=" * 100)
            
            print("{} {:>30} {:>47}".format(
                "Order ID: " + orders["order id"],
                "Table Number: " + str(orders["table number"]),
                "Date: " + orders["date"]
            ))
            
            print("-" * 100)
            
            tot_amt = 0
            for key in orders:
                if key in ("starters", "dinner", "desserts") and orders[key]:
                    print(">> {}".format(key.capitalize()))
                    print("." * 100)
                    print("{:<8} {:<25} {:<10} {:<10} {:>15}".format(
                        "Code", "Name", "Price", "Quantity", "Total Price"
                    ))
                    
                   
                    for inner_di in orders[key]:
                        tot_amt += (inner_di["price"] * inner_di["quantity"])
                        print("{:<8} {:<25} {:<10} {:<10} {:>10}".format(
                            inner_di["code"],
                            inner_di["name"],
                            inner_di["price"],
                            inner_di["quantity"],
                            str(inner_di["price"] * inner_di["quantity"])
                        ))
                    print("-" * 100)
                    
                
            print("{} {}".format("TOTAL AMOUNT: ", str(tot_amt)))
            print("=" * 23)
        except Exception as e:
            print(e)
        
        Orders.tables_reserved.remove(orders["table number"])
        
    def save_bill(self, orders):
        try:
            file_name = "ORD0" + str(orders["table number"]) + "_" + str(datetime.now().strftime("%I-%M"))
            
            file_path = r"C:\Users\Sudip\Downloads\Bizmetric Training\Python\Bills" + f"\{file_name}"
            
            with open(file_path, "w") as file:
                file.write("=" * 100,)
                file.write("\n")
                file.write("|{:^98}|".format("BILL"))
                file.write("\n")
                file.write("=" * 100)
                
                file.write("\n")
                file.write("{} {:>30} {:>47}".format(
                    "Order ID: " + orders["order id"],
                    "Table Number: " + str(orders["table number"]),
                    "Date: " + orders["date"]
                ))

                file.write("\n")
                file.write("-" * 100)

                tot_amt = 0
                for key in orders:
                    if key in ("starters", "dinner", "desserts") and orders[key]:
                        file.write("\n")
                        file.write(">> {}\n".format(key.capitalize()))
                        file.write("." * 100)

                        file.write("\n")
                        file.write("{:<8} {:<25} {:<10} {:<10} {:>15}\n".format(
                            "Code", "Name", "Price", "Quantity", "Total Price"
                        ))

                        for inner_di in orders[key]:
                            tot_amt += (inner_di["price"] * inner_di["quantity"])
                            file.write("\n")
                            file.write("{:<8} {:<25} {:<10} {:<10} {:>10}".format(
                                inner_di["code"],
                                inner_di["name"],
                                inner_di["price"],
                                inner_di["quantity"],
                                str(inner_di["price"] * inner_di["quantity"])
                            ))

                file.write("\n")
                file.write("-" * 100)
                        
                file.write("\n")
                file.write("{} {}".format("TOTAL AMOUNT: ", str(tot_amt)))
                file.write("\n")
                file.write("=" * 23)
        except Exception as e:
            print(e)
```
