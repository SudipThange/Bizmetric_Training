***Comprehensive list:***

```python
# Q1. Write a list comprehension to generate squares of numbers from 1 to 10.

squares = [i ** 2 for i in range(1, 11)]

print("SQUARE FROM 1 TO 10: ", squares)
```

```python
# Q2. Create a list of even numbers between 1 and 50 using list comprehension.

even_nums = [i for i in range(1, 51) if i % 2 == 0]
print("EVEN NUMBER: ", even_nums)
```

```python
# Q3. Convert all strings in a list to uppercase using list comprehension

text = ["Bat", "Animal", "Airline", "Truck"]

names_upper = [s.upper() for s in text]
print("TEXT IN UPPER CASE: ", names_upper)
```

```python
# Q4. Given a list of integers, create a new list that contains only the positive numbers.

nums = [1, -1, 2, 5, 10, 2, -3, -7, -2, 8]

positive_nums = [i for i in nums if i > 0]
print("POSITIVE NUMBERS ARE: ", positive_nums)
```

```python
# Q5. Create a list of tuples (num, num^2) for numbers 1 to 5.

li = [(i, i ** 2) for i in range(1, 6)]
print(li)
```

```python
# Q6. Extract all vowels from a given string using list comprehension.

text = input("Enter the string?: ")

if not text.replace(" ", "").isalpha():
    print("PLESE RE-ENTER STRING AGAIN...")
else:
    vowels = [ch for ch in text if ch in ('a', 'e', 'i', 'o', 'u')]
    print("VOWELS FOR [{}] ARE: {}".format(text, set(vowels)))
```

```python
# Q7. Flatten a 2D list using list comprehension.

li = [[10, 20], [30, 40]]

flatten_li = [j for i in li for j in i]
print("FLATTEN ARRAY: ", flatten_li)
```

```python
# Q8. Replace all negative numbers in a list with 0 using list comprehension.

nums = [1, -1, 2, 5, 10, 2, -3, -7, -2, 8]

nums = [0 if i < 0 else i for i in nums]
print("LIST AFTER NEGATIVE REPLCE: ", nums)
```

```python
# Q11.  From a list of numbers, generate a list of â€œevenâ€ or â€œoddâ€ strings using list comprehension

li = [i for i in range(1, 11)]

filtered_li = ["even" if i % 2 == 0 else "odd" for i in li]

for i, text in enumerate(filtered_li, start = 1):
    print(f"{i} : {text}", end = ", ")
```

```python
# Q1. Write a lambda to add two numbers.

def get_numbers():
    num1 = int(input("Enter the first number?: "))
    num2 = int(input("Enter the second number?: "))
    
    return num1, num2

add = lambda a,b: a + b

a, b = get_numbers()
res = add(a, b)
print("Addition: ", res)
```

```python
# Q2. Create a lambda to check if a number is even.

number = int(input("ENTER THE NUMBER?: "))
print("NUMBER IS: ", number)

check_even = (lambda x: x % 2 == 0)

if check_even(number):
    print("NUMBER IS EVEN!!!")
else:
    print("NUMBER IS ODD!!!")
```

```python
# Q3. Write a lambda to get the last character of a string.

text = input("Enter the text?: ")
print("Text is: ", text)

get_last_char = (lambda text: text[-1])

last_char = get_last_char(text)
print(f"LAST CHAR IS: {last_char}")
```

```python
# Q4. Use lambda with map() to square every number in a list.

li = [elem for elem in range(1, 11)]
print("LIST ELEMENTS: ", li)

square = (lambda x: x**2)

square_list = list(map(square, li))
print("SQAURE OF THE LIST ELEMENTS: ", square_list)
```

```python
# Q5. Use lambda with filter() to get only odd numbers from a list.

li = [elem for elem in range(1,11)]
print("LIST ELEMENTS: ", li)

filter_numbers = (lambda x: x % 2 != 0)

odd_numbers = list(filter(filter_numbers, li))
print("ODD NUMBERS: ", odd_numbers)
```

```python
# Q6. Use sorted() + lambda to sort a list of tuples by second value.

li = [(1, 3), (2, 5,), (10, 1), (5, 4)]
print("LIST: ", li)

func = (lambda x: x[1])

sorted_by_second_index = sorted(li, key = func)
print("Sorted list: ", sorted_by_second_index)
```

```python
# Q7. Create a lambda to check if a string is a palindrome.

txt = input("Enter the text: ").lower()
print("Text is: ", txt)

check_palindrome = (lambda txt: txt[::-1])

if txt == check_palindrome(txt):
    print("PALINDROME STRING!!")
else:
    print("NOT PALINDROME STRING!!")
```

```python
# Q27. Write a lambda to check if two strings are anagrams.

text_1 = input("Enter the first string?: ").lower()
text_2 = input("Enter the second string?: ").lower()
print("Entered strings are: ", text_1, text_2)

check_anagram = (lambda s1, s2: True if "".join(sorted(s1)) == "".join(sorted(s2)) else False)

if check_anagram(text_1, text_2):
    print("YESS, ANAGRAMS!!")
else:
    print("NOO, ANAGRAMS!!")
```

```python
# Q29. Use lambda inside any() to check if any list element is negative.

nums = [1, 2, 3, 4, 5, 10]

check_negative = (lambda n: True if n < 0 else False)

if any([check_negative ,nums]):
    print("List contains a negative number...")
else:
    print("List does not contains any negative number...")
```
