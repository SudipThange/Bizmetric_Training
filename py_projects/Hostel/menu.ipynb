```python
from admin import Admin


curr_admin = None

while True:
    print()
    print("*" * 100)
    print("|{:^98}|".format("ADMIN OPTIONS"))
    print("*" * 100)
    print("| Enter 1:  Add Admin. {:>77}".format("|"))
    print("| Enter 2:  Show Admin. {:>76}".format("|"))
    print("| Enter 3:  Login. {:>81}".format("|"))
    print("| Enter 4:  Add Students. {:>74}".format("|"))
    print("| Enter 5:  Show Students. {:>73}".format("|"))
    print("| Enter 6:  Genrate Fee Recipt. {:>68}".format("|"))
    print("| Enter 7:  Print Fee Recipt. {:>70}".format("|"))
    print("| Enter 8:  Logout. {:>80}".format("|"))
    print("| Enter 9:  Exit. {:>82}".format("|"))
    print("-" * 100)
    choice = int(input(">> Enter your choice?: "))
    
    if curr_admin is None:
        curr_admin = Admin()
        
    if choice == 1:
        curr_admin.register()
        
    elif choice == 2:
        curr_admin.show_all_admins()
        
    elif choice == 3:
        curr_admin.login()
        
    elif choice == 4:
        curr_admin.add_students()
        
    elif choice == 5:
        curr_admin.show_all_students()
        
    elif choice == 6:
        curr_admin.print_on_console()
        
    elif choice == 7:
        curr_admin.save_to_file() 
        
    elif choice == 8:
        curr_admin.logout()

    elif choice == 9:
        print("Exiting...")
        break
```
